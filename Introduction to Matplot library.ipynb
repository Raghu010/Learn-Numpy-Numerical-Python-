__Matplot-Library__: Is a python package which is used to visualize the 2D(dimensional) data.

This library imports most of the functionalities from mat-lab.

```python
#importing the library.
import matplotlib.pyplot as plt
```

__py__ extension is used for python mat-lab functionalities.

Note: C, C++, Java also imports the visualization tools from mat-lab so we use py extension.

```python
#example visualizing an array.
import numpy as np
a = np.array([1, 2, 3, 4, 5])
plt.plot(a)
plt.show() #this line of code is compulsory if you are executing in python IDLE.(But not required in ipython)
```

```python
#examples using matplot library.
import numpy as np
import matplotlib.pyplot as plt
a = np.array([1, 3, 5, 2, 8])
plt.plot(a)
plt.show()
```

```python
#plot for array of zeros.
a = np.zeros(5)
plt.plot(a)
plt.show()
```

```python
#plot for array of ones.
a = np.ones(5)
plt.plot(a)
plt.show()
```

```python
#plot for array with junk and zero values.
a = np.empty(8)
plt.plot(a)
plt.show()
```

```python
#example.
a = np.array([1, 2, 3])
plt.scatter(a, np.sin(a))
plt.plot(a, np.sin(a))
plt.scatter(a, np.sin(a))
```

```python
#example for a scalar value.
a = np.array(5)  #Note: There will be no output for a scalar value.
plt.plot(a)
plt.show()
```

```python
#plot using shape method.
a = np.random.rand(10)
a.shape = 5, 2
plt.plot(a)
plt.show()
```

# Plotting the Trigonometric functions.

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.linspace(0, 2*np.pi, 10)
b = np.sin(a)
plt.plot(b)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.linspace(0, 2*np.pi, 10)
b = np.cos(a)
plt.plot(b)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.linspace(0, 2*np.pi, 10)
b = np.tan(a)
plt.plot(b)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.linspace(0, 2*np.pi, 10)
b = (1/(np.sin(a)))
plt.plot(b)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.linspace(0, 2*np.pi, 10)
b = (1/(np.cos(a)))
plt.plot(b)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.linspace(0, 2*np.pi, 10)
b = (1/(np.tan(a)))
plt.plot(b)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.random.random(10)
plt.hist(a)
plt.show() #note: This function generates the bar plot.
```
